#### Create all possible permutations of parameters for all but SAME templates

```python
all_modifiers = []
for modifier_type in param_types["Modifier"]:
    for modifier_key in param_types["Modifier"][modifier_type]:
        #if modifier_key in synonyms:
        #    words = synonyms[modifier_key]
        #else:
        words = [modifier_key]
        for word in words:
            all_modifiers.append(word)
```

```python
all_modifiers
```

```python
all_referents = []
for modifier in all_modifiers:
    if modifier != '':
        #for word in synonyms['thing']:
        all_referents.append([modifier, 'thing'])
    for shape_type in param_types["Noun"]["Shape"]:
        #if shape_type in synonyms:
        #    words = synonyms[shape_type]
        #else:
        words = [shape_type]
        for word in words:
            all_referents.append([modifier, word])
            
    
```

```python
all_referents
```

```python
all_param_combinations = []

def combine_ref1_ref2(referent1, referent2):
    #if referent1[0] in synonyms:
    #    referent10 = synonyms[referent1[0]]
    #else:
    referent10 = [referent1[0]]
    #if referent1[1] in synonyms:
    #    referent11 = synonyms[referent1[1]]
    #else:
    referent11 = [referent1[1]]
    #if referent2[0] in synonyms:
    #    referent20 = synonyms[referent2[0]]
    #else:
    referent20 = [referent2[0]]
    #if referent2[1] in synonyms:
    #    referent21 = synonyms[referent2[1]]
    #else:
    referent21 = [referent2[1]]
    for p0 in referent10:
        for p1 in referent11:
            for p2 in referent20:
                for p3 in referent21:
                    all_param_combinations.append([p0, p1, p2, p3])


for referent1 in all_referents:
    for referent2 in all_referents:
        if referent1[0] != referent2[0]:
            combine_ref1_ref2(referent1, referent2)
        elif referent1[1] != 'thing' and referent2[1] != 'thing' and referent1[1] != referent2[1]:
            combine_ref1_ref2(referent1, referent2)            
```

#### Create all possible permutations of parameters for SAME template

```python
all_modifiers = []
for modifier_type in param_types["Modifier"]:
    for modifier_key in param_types["Modifier"][modifier_type]:
        #if modifier_key in synonyms:
        #    words = synonyms[modifier_key]
        #else:
        words = [modifier_key]
        for word in words:
            all_modifiers.append((word, modifier_type))
```

```python
all_modifiers
```

```python
all_referents = []
for modifier in all_modifiers:
    if modifier[1] != 'Null':
        #for word in synonyms['thing']:
        all_referents.append([modifier, 'thing'])
    for shape_type in param_types["Noun"]["Shape"]:
        #if shape_type in synonyms:
        #    words = synonyms[shape_type]
        #else:
        words = [shape_type]
        for word in words:
            all_referents.append([modifier, word])
```

```python
all_referents
```

```python
all_combinations = []

property_types = list(param_types["Property"].keys())
for referent in all_referents:
    if referent[1] == 'thing':
        exclude_property = [referent[0][1]]
        properties = [p for p in property_types if p not in exclude_property]
        for p in properties:
            all_combinations.append([referent[0][0], referent[1], param_types["Property"][p]])
    else:
        exclude_property = [referent[0][1], 'Shape']
        properties = [p for p in property_types if p not in exclude_property]
        for p in properties:
            all_combinations.append([referent[0][0], referent[1], param_types["Property"][p]])
```

```python
all_combinations
```

```python
all_param_combinations = []

for m,n,p in all_combinations:
    #if m in synonyms:
    #    referent_m = synonyms[m]
    #else:
    referent_m = [m]
    #if n in synonyms:
    #    referent_n = synonyms[n]
    #else:
    referent_n = [n]
    #if p in synonyms:
    #    referent_p = synonyms[p]
    #else:
    referent_p = [p]
    for p0 in referent_m:
        for p1 in referent_n:
            for p2 in referent_p:
                all_param_combinations.append([p0,p1,p2])
```
